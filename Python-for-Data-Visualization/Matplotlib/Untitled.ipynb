```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y)
ax2.plot(x,y)
```

```python
fig = plt.subplots(2,1)
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,0.4])

ax1.plot(x,z)
ax1.set_xlim([0,100])
ax1.set_ylim([0,10000])

ax2.plot(x,y)
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
plt.subplots(1,2)
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,2))

axes[0].plot(x,y,ls='--',lw=3, alpha = 1)
axes[1].plot(x,z, color = 'red', lw = 3)
```
